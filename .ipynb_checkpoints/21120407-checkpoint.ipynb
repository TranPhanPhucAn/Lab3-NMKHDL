### Requirement #4: Relationship between 2 variables

- Exploring the relationship between variables is a needed task to do
- You will start with exploring the linear relationship between 2 variables. In order to do so, draw $C_6^2=15$ scatter plots. For each plot, draw a line that shows the linear relationship of variables. You should implement the line yourself instead of using library.

````python
# TODO: scatter stuff
n=len(df['Total Cases'])
def create_relationship(axis,col1,col2):
    x=df[col1].copy()
    y=df[col2].copy()
    
    m=(n*(x*y).sum()-x.sum()*y.sum())/(n*(x**2).sum()-(x.sum())**2)
    b=(y.sum()-m*(x.sum()))/n
    #create basic scatterplot
    axis.plot(x, y, 'o')
    #add linear regression line to scatterplot

    axis.plot(x, m*x+b,color='red')
    axis.set_xlabel(col1)
    axis.set_ylabel(col2)
    axis.set_title(f'{col1}-{col2}')

fig, axes = plt.subplots(5, 3, figsize=(17, 20))
row=0
col=0
index1=0
index2=1
for i in range(15):
    create_relationship(axes[row, col], fields[index1],fields[index2])
    if col==2:
        row+=1
        col=0
    else:
        col+=1
    if index2==5:
        index1+=1
        index2=index1+1
    else:
        index2+=1

# Adjust layout
plt.tight_layout()
plt.show()
````

### Requirement #5: Double check the relationship between 2 variables

- In order to make sure about the linear relationship, you have to check each pair with **Pearson correlation coefficient**
- Specifically, $\rho$ is calculated as follows:

$$\rho_{X, Y} = \frac{cov(X, Y)}{\sigma_X \sigma_Y}$$

- Then, plot these 15 values using heatmap

````python
def calculate_pearson_correlation(df, col1,col2):
    cov_xy = ((df[col1] - df[col1].mean()) * (df[col2] - df[col2].mean())).sum() / (len(df) - 1)
    std_x = df[col1].std()
    std_y = df[col2].std()
    return cov_xy/(std_x*std_y)
columns = fields
correlation_matrix = pd.DataFrame(index=columns, columns=columns)

for i in range(len(columns)-1):
    for j in range(i+1,len(columns)):
        correlation_matrix.loc[columns[j], columns[i]] = calculate_pearson_correlation(df, columns[j], columns[i])
correlation_matrix = correlation_matrix.apply(pd.to_numeric, errors='coerce')
plt.figure(figsize=(8, 6))
sn.heatmap(correlation_matrix, annot=True, fmt=".2f", linewidths=.5,vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
````

### Requirement #7: Create models

- Create 2 models as mentioned above. Call them `Mnist_FC` and `Mnist_CNN`
- **I strongly recommend that you should slowly implement, print out the output and its shape of each layer in order to get a deep understanding of what you are trying to do**
- Before moving to the next requirement, you should test your model with a random sample of the dataset. In case you create code cells for testing, remember to remove these cells

````python
# TODO: create Mnist_FC and Mnist_CNN
class Mnist_FC(nn.Module):
    def __init__(self):
        super(Mnist_FC, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(784, 1024)
        self.relu1 = nn.ReLU()
        
        self.fc2 = nn.Linear(1024, 512)
        self.relu2 = nn.ReLU()
        
        self.fc3 = nn.Linear(512, 10)
        self.logsoftmax = nn.LogSoftmax(dim=1)
    def forward(self,x):
        # print(x.size())
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        x = self.logsoftmax(x)
        return x

class Mnist_CNN(nn.Module):        
    def __init__(self):
        super(Mnist_CNN, self).__init__()

        # First convolutional layer
        self.conv1 = nn.Conv2d(1, 20, kernel_size=5, padding=2)  # Assuming input size is 3 channels (e.g., RGB image)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        # Second convolutional layer
        self.conv2 = nn.Conv2d(20, 20, kernel_size=5, padding=2)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Fully connected layer
        self.fc = nn.Linear(20 * 7 * 7, 10)  # Adjust the input size based on your input image size
        self.logsoftmax = nn.LogSoftmax(dim=1)

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        # Flatten the output for the fully connected layer
        x = x.reshape(x.size(0), -1)
        x = self.logsoftmax(self.fc(x))
        return x
````

### Requirement #8: Train and test models

- Design `train` and `test` function with the parameters provided in the following code cell
    - `train` function trains the model with all the data in training set
    - `test` function tests the model with all data in testing set

- Return the accuracy (e.g. $0.847351$) after training and testing

**TODO**: Explain the parameters of `train` and `test` function. In `test` function, there is a line of code that says `with torch.no_grad()`. Why do we have this line of code?


train funtion:
- dataloader: This parameter typically refers to a data loader object. Training data is often too large to fit into memory at once. A data loader is an iterable that allows you to load batches of data sequentially, making it memory-efficient.
- model: his parameter represents the machine learning model that you want to train. In this assignment, it's neural network and full connection.
- loss_fn: This parameter refers to the loss function. The loss function measures the difference between the model's predictions and the actual target values (labels). During training, the goal is to minimize this loss.
- optimizer: This parameter typically represents the optimization algorithm used to update the model parameters during training.

test function:
-  dataloader: This parameter typically refers to a data loader object. It's used to iterate through batches of test data.
-  model: his parameter represents the machine learning model that you want to train. In this assignment, it's neural network and full connection.

-  with torch.no_grad():  you prevent PyTorch from tracking operations for gradient computation within the associated code block. The program explicitly uses it here (as with most neural networks) in order to not update the gradients when it is updating the weights as that would affect the back propagation. Disabling gradient computation can also be beneficial in terms of computational efficiency.

````python
# TODO: design train and test function

# check gpu
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f'Using {device} for training and testing')

def train(dataloader, model, loss_fn, optimizer):
    model
    total_step = len(dataloader)
    correct = 0
    total=0
    for i, (images, labels) in enumerate(dataloader):
        images = images.to(device)
        labels = labels.to(device)
            
        # Forward pass
        outputs = model(images)
        loss = loss_fn(outputs, labels)
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    print('Training acc: {} '.format(correct / total))
    return correct / total
def test(dataloader, model):
    model.eval()  # eval mode (batchnorm uses moving mean/variance instead of mini-batch mean/variance)
    with torch.no_grad():
        correct = 0
        total = 0
        for images, labels in dataloader:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
        print('Testing acc: {} '.format(correct / total))
        return correct / total
````

````python
# define model
model_CNN = Mnist_CNN().to(device)
model_FC = Mnist_FC().to(device)
loss_fn = nn.CrossEntropyLoss()
learning_rate = 1e-3
optimizer_CNN = torch.optim.SGD(model_CNN.parameters(), lr=learning_rate)
optimizer_FC = torch.optim.SGD(model_FC.parameters(), lr=learning_rate)
history_fc={}
history_cnn={}
# TODO: run 20 epochs for Mnist_FC and Mnist_CNN
print('========== FC model ==========')

for epoch in range(num_epochs):
    print(f'\nEpoch {epoch+1}\n----------------------\n')
    acc_train = train(train_loader, model_FC, loss_fn, optimizer_FC)
    acc_test = test(test_loader, model_FC)
    history_fc[epoch]=[acc_train,acc_test]
    
print('========== CNN model ==========')
for epoch in range(num_epochs):
    print(f'\nEpoch {epoch+1}\n----------------------\n')
    acc_train = train(train_loader, model_CNN, loss_fn, optimizer_CNN)
    acc_test = test(test_loader, model_CNN)
    history_cnn[epoch]=[acc_train,acc_test]
    
````
